```python
import pandas as pd
from textblob import TextBlob
import matplotlib.pyplot as plt

data = pd.read_csv('twitter_training.csv')
data.head()
```

```python
col_names = ['ID', 'Place', 'Emotion','Content']
df = pd.read_csv('twitter_training.csv', names=col_names)
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
#to check how many null values exist
df.isnull().sum()
```

```python
df.dropna(axis=0, inplace=True)
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
df.duplicated().sum()
```

```python
df.shape #after deleteing the null and duplicated values we check the size
```

```python
sentiment_counts = df['Emotion'].value_counts()
sentiment_counts
```

```python
#bar plot
plt.figure(figsize=(8,6))
sentiment_counts.plot(kind='bar', color=['red', 'blue', 'purple', 'yellow'])
plt.title('Sentiment Counts')
plt.xlabel('Sentiment')
plt.xticks(rotation=0)  # displays the values i.e 'positive' in horizontal way
plt.ylabel('Count')
plt.show()
```

```python
brand_data = df[df['Place'].str.contains('Microsoft', case = False)]
brand_sentiment_counts = brand_data['Emotion'].value_counts()
brand_sentiment_counts
```

```python
#pie chart
plt.figure(figsize=(6,6))
plt.pie(brand_sentiment_counts, labels=brand_sentiment_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Sentiment Distribution')
plt.show()

```
